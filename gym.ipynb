```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your dataset
# Replace 'your_dataset.csv' with the path to your dataset
# data = pd.read_csv('your_dataset.csv')

# Set up the scatter plot
plt.figure(figsize=(10, 6))

# Choose two columns to plot
# For example, you can plot 'Calories_Burned' vs 'Session_Duration (hours)'
x = 'Session_Duration (hours)'
y = 'Calories_Burned'

# Create the scatter plot
sns.scatterplot(data=has, x=x, y=y, hue='Workout_Type', style='Experience_Level', s=100)

# Customize the plot
plt.title('Scatter Plot of Calories Burned vs Session Duration')
plt.xlabel(x)
plt.ylabel(y)
plt.legend(title='Workout Type and Experience Level')
plt.grid(True)

# Show the plot
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load your dataset
# Replace 'your_dataset.csv' with the path to your dataset
# data = pd.read_csv('your_dataset.csv')

# Set up the count plot
plt.figure(figsize=(10, 6))

# Choose the categorical variable for the count plot
# For example, if you want to plot the frequency of 'Workout_Type'
sns.countplot(data=has, x='Workout_Type', palette='Set2')

# Customize the plot
plt.title('Count of Workout Types')
plt.xlabel('Workout Type')
plt.ylabel('Count')

# Show the plot
plt.xticks(rotation=45)  # Rotate x-axis labels if necessary
plt.tight_layout()
plt.show()
```

```python
sns.violinplot(x="Max_BPM",y="Avg_BPM", data=has)
plt.figure(figsize=(30,30))
plt.show()
```

```python
has.columns
```

```python
# Box plot for Fat Percentage by Workout Type
sns.boxplot(data=has, x='Workout_Type', y='Fat_Percentage', palette='Set2')
plt.title('Fat Percentage by Workout Type')
plt.xlabel('Workout Type')
plt.ylabel('Fat Percentage')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Pair plot for select numeric variables
sns.pairplot(has[['Age', 'Weight (kg)', 'Height (m)', 'Calories_Burned', 'Fat_Percentage']])
plt.suptitle('Pairwise Relationships', y=1.02)
plt.show()
```
